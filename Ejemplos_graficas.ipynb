##Gráfica de la función exponencial (Ejemplo)

```python
#Cargar la librería matplotlib
import matplotlib.pyplot as plt
import numpy as np
#Definir los límites de x
x=np.linspace(0,10,100)
#Definir la función en y 
y=np.exp(-x)
plt.plot(x,y,color="green")
plt.xlabel("valores de x")
plt.ylabel("$f(x)$")
plt.title("$f(x)=e^{-x}$")
plt.show
```

```python
#Definir los límites de x
x=np.linspace(0,10,100)
#Definir la función en y 
y=np.exp(2*x)
plt.plot(x,y)
plt.title("$g(x)=e^{2x}$")
plt.xlabel("valores de x")
plt.ylabel("$g(x)$")
plt.show
```

**Función seno y coseno**

```python
#gráficas y plotting
import numpy as np
import matplotlib.pyplot as plt
#Definir las coordenadas en x e y
x=np.arange(0,10*np.pi,0.1)
y=np.cos(x)
z=np.sin(x)
#plot usando matplotlib
plt.plot(x,y)
plt.plot(x,z)
plt.xlabel("valores de x")
plt.title("sen(x) y cos(x)")
plt.show()
```
